#### Random Forest Practice

```python
import pandas as pd

melb_file_path = '../../Datasets/melb_data.csv'
melb_data = pd.read_csv(melb_file_path)

melb_data = melb_data.dropna(axis=0)
melb_data.describe()
```

```python
y = melb_data.Price

melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea', 'YearBuilt', 'Lattitude', 'Longtitude']

X = melb_data[melbourne_features]

from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)

melb_preds = forest_model.predict(val_X)
melb_mae = mean_absolute_error(val_y, melb_preds)

print("MAE for Random Forest Model: {:,.0f}".format(melb_mae))
```
