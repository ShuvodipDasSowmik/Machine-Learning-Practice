### String Functions

```python
phrase = "Amar Sonar Bangla"

print(phrase.capitalize())
print(phrase.upper())
print(phrase.isupper())
print(phrase[0])

print(len(phrase))

print(phrase.replace("Bangla", "London"))

# Beginning to 2nd index
print(phrase[:3])

# 4th index to end
print(phrase[4:])

# 5th index to 12th index
print(phrase[5:13])

"""
Syntax: phrase[start:stop:step]

start = 2 → include character at index 2
stop = 15 → go up to but not including index 15
step = 2 → take every 2nd character (stride 2)
So Python picks indices: 2, 4, 6, 8, 10, 12, 14.
"""
print(phrase[2:15:2])
```

```python
print(phrase[::2])

# Reverse of the string
print(phrase[::-1])

# Reverse of phrase[::2]
print(phrase[::-2])
```

```python
# index: -x means xth index from the end

print(phrase[-1])

print(phrase*2)
```

```python
# Works same for lists

listx = [1, 2, 3, 4, 5]

print(listx[::-1])
```

### Arithmetic Operations

```python
i, j, k = 5, 10.5, 3

sum = i + j + k
diff = j - i
print(f"Sum: {sum}, Difference: {diff}")

mul = i*j
div = i/j
print(f"Multiplication: {mul}, Division: {div}")

exp = i**j
print("Exponential of i to j", exp)
```

```python
# integer division

idiv = i//k

print("Integral Division:", idiv)

# i++ or i-- isn't defined in python, however the following are
i = 5
print("I =", i)
i += 3
i %= 3

print("I =", i)

i **= 3;
print("I =", i)
```

### Bitwise Operation

```python
a, b = 7, 14

# a = 0111, b = 1110


print(a & b)    # a & b = 0110 -> 6

print(a | b) # a | b = 1111 -> 15

print(a^b) # a ^ b = 1001 -> 9

print(~b)   # b = 0000 1110 ~b = 1111 0001

print(a << 1) # a << 1 = 1110
print(a >> 1) # a >> 1 = 0011

print(bin(123))
```

### Logical Operator: and, not, or

```python
x = int(input())

if x == 2 or x == 3:
    print("X is 2 or 3")
elif x == 5:
    print("X is 5")
else:
    print("X is unknown")
```

#### Membership Operation: in, not in

#### Identity Operation: is, is not

```python
l1 = [1, 2, 3, 5]
l2 = [1, 2, 3, 5]
l3 = [12, 12, 1, 1]

print(2 in l1)
print(5 not in l2)
print(l1 is l2)
print(l1 is l3)
print(l1 is not l3)
```

#### Loops and Ranges

```python
list2 = list(range(1, 23))

# range(m, n) returns m to n-1

list2
```

```python
i = 0

while i < 10:
    print(i, end=' ')
    i += 1
```

```python
# Notice here 'in' is a membership operator

for i in range(1, 10):
    print(i, end=' ')
    i += 1
```

```python
fruits = ['Banana', 'Pineapple', 'Aam']

for fruit in fruits:
    print(fruit, end=' ')

print()

for index in range(0, len(fruits)):
    print(fruits[index], end=' ')
```

```python
num = [1, 3, 7, 5, 9, 11]

for n in num:
    if(n % 2 == 0):
        print("Even number found")
        break;
else:
    # Executed when loop is completed without hitting break
    print("No even number found")
```
