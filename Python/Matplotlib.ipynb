```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 20, 20)
print(x)

y = np.sin(x)
print(y)
```

```python
y2 = np.cos(x)
```

```python
# Bare minimum

plt.plot(x, y)
plt.show()
```

```python
plt.figure(figsize=(4, 2)) # default 8,6
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, label='sin(x)', color='red', linestyle='dashdot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
plt.plot(x, y, label='sin(x)', color='red', linestyle='dashdot')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y2, label='cos(x)', color='orange', linestyle='dashed')
plt.legend()    # Legend shows the data passed in the label param of plt.plot
plt.show()
```

```python
plt.plot(x, y, label='sin(x)', color='red', linestyle='dashdot')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y2, label='cos(x)', color='orange', linestyle='dashed')
plt.plot(x, y*y2, label='sin(x)cos(x)', color='blue', linestyle='dotted')
plt.legend()    # Legend shows the data passed in the label param of plt.plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)', color='red', linestyle='dashdot')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y2, label='cos(x)', color='orange', linestyle='dashed')
plt.plot(x, y*y2, label='sin(x)cos(x)', color='blue', linestyle='dotted')
plt.legend(loc='upper right')   # Specify the location of legend
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)', color='red', linestyle='dashdot')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y2, label='cos(x)', color='orange', linestyle='dashed')
plt.plot(x, y*y2, label='sin(x)cos(x)', color='blue', linestyle='dotted')
plt.legend(loc='upper right')
plt.grid(True)  # Shows grid in the graph
plt.title('Trigonometric Functions Relations')
plt.show()
```
