```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a, b = 0, 1
N = 10

dx = (b-a)/N

x = np.linspace(a, b, N+1)
y = x**2

print(x)
print(y)
```

```python
y_left = y[1:]
print(y_left)

y_right = y[:-1]
print(y_right)
```

```python
integral = np.sum((dx/2)*(y_left + y_right))

integral
```

```python
plt.figure(figsize=(4,2))
plt.plot(x, y)
plt.grid(True)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# Or more easily, use the trapz() function

A = np.trapezoid(y, x, dx)
print(A)
```
