```python
from abc import ABC, abstractmethod
```

```python
class Animal(ABC):
    def __init__(self, species):
        self.species = species
    
    @abstractmethod
    def make_sound():
        pass

    def get_species(self):
        return self.species

class Dog(Animal):

    breed_count = 0

    def __init__(self, breed):
        super().__init__("Mammal")
        self.breed = breed
        Dog.breed_count += 1

    def make_sound(self):
        print("Bark! Bark!")

    def get_breed(self):
        return self.breed;

    @staticmethod
    def get_breed_count():
        return Dog.breed_count

class Corgi(Dog):
    def __init__(self, name):
        super().__init__("Corgi")
        self.name = name

    def walk(self):
        print("Walking with small legs")
```
