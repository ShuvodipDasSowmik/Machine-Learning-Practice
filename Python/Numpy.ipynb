```python
# np.concat((arr, arr2))
# Must have same number of dimensions

arr0 = np.random.randint(1, 100, (3, 3))
arr01 = np.random.randint(1, 100, (3, 3))

arr3 = np.concat((arr0, arr01))
print(arr3, end='\n\n')
print(arr3.ndim)
print(arr3.shape, end='\n\n')

print(arr3.transpose(), end='\n\n')
print(np.sort(arr3), end='\n\n')
print(np.sort(arr3.transpose()))

# All entries aren't sorted, instead each list under the matrix is sorted
```

```python
predictions = [1, 2, 4, 5, 1, 3, 2, 4, 1.2, 0]
label = [0, 3, 1, 2, 3, 1, 5, 6, 7 , 2]

npreds = np.array(predictions)
nlabel = np.array(label)

n = len(predictions)

avg = (1/n) * np.sum(np.square(npreds - nlabel))

avg
```
